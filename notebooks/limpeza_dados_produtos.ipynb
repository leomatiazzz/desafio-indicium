# Normalização de dados de produtos

Este notebook realiza a normalização dos dados presentes no arquivo `produtos_raw.csv`, conforme as tarefas obrigatórias:

1. Padronizar os nomes das categorias de produtos em: eletrônicos, propulsão e ancoragem.
2. Converter os valores para o tipo numérico.
3. Remover as duplicatas.

Utilizamos apenas o CSV `produtos_raw.csv` e Python 3.

```python
import csv
```

## Padronização das categorias

Definimos uma função para normalizar as categorias, removendo espaços, acentos e padronizando para "eletrônicos", "propulsão" ou "ancoragem".

```python
def normalize_category(cat):
    cat_lower = cat.lower().replace(' ', '').replace('ô', 'o').replace('ã', 'a').replace('ç', 'c')
    if 'eletr' in cat_lower:
        return 'eletrônicos'
    elif 'propul' in cat_lower:
        return 'propulsão'
    elif 'ancor' in cat_lower:
        return 'ancoragem'
    else:
        return cat
```

## Leitura e processamento do CSV

Lemos o arquivo `produtos_raw.csv`, aplicamos a normalização das categorias e convertemos os preços para float.

```python
data = []
with open('../datasets/produtos_raw.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        row['actual_category'] = normalize_category(row['actual_category'])
        row['price'] = float(row['price'].replace('R$ ', '').replace(',', ''))
        data.append(row)

print(f"Total de linhas lidas: {len(data)}")
```

## Remoção de duplicatas

Removemos linhas duplicadas com base em todos os campos.

```python
seen = set()
unique_data = []
for row in data:
    row_tuple = tuple(row.values())
    if row_tuple not in seen:
        seen.add(row_tuple)
        unique_data.append(row)

print(f"Duplicatas removidas: {len(data) - len(unique_data)}")
print(f"Total de linhas após limpeza: {len(unique_data)}")
```

## Salvamento dos dados limpos

Salvamos o resultado em `produtos_clean.csv`.

```python
with open('../features/produtos_clean.csv', 'w', newline='', encoding='utf-8') as f:
    if unique_data:
        fieldnames = unique_data[0].keys()
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(unique_data)

print("Dados normalizados salvos em features/produtos_clean.csv")
```

## Resumo

Este notebook completou a normalização dos dados:

- **Categorias padronizadas**: Todas as variações foram unificadas em "eletrônicos", "propulsão" e "ancoragem".
- **Preços convertidos**: De strings formatadas para valores numéricos (float).
- **Duplicatas removidas**: Linhas idênticas foram eliminadas.

O arquivo `produtos_clean.csv` está pronto para análise posterior.
