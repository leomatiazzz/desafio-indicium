# Normalização de custos de importação
Este notebook converte `custos_importacao.json` para um CSV plano com `product_id`, `start_date` (YYYY-MM-DD) e `usd_price`.

```python
import pandas as pd
import json
```

## Carregar e transformar JSON de custos
- Cada item de `historic_data` vira uma linha no DataFrame
- Datas serão normalizadas para `YYYY-MM-DD`

```python
with open('../datasets/custos_importacao.json', 'r', encoding='utf-8') as f:
    custos_data = json.load(f)

rows = []
for item in custos_data:
    product_id = item.get('product_id')
    for historic in item.get('historic_data', []):
        start_date = pd.to_datetime(historic.get('start_date'), dayfirst=True, errors='coerce')
        rows.append({
            'product_id': product_id,
            'start_date': start_date,
            'usd_price': historic.get('usd_price')
        })

custos_df = pd.DataFrame(rows)
custos_df = custos_df.dropna(subset=['product_id', 'start_date', 'usd_price'])
custos_df['start_date'] = custos_df['start_date'].dt.date
print(custos_df.head())
print('Total de linhas processadas:', len(custos_df))
```

## Salvar CSV normalizado

```python
custos_df.to_csv('../features/custos_importacao_normalizado.csv', index=False)
print('Arquivo salvo: features/custos_importacao_normalizado.csv')
```
