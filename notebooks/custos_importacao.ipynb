# Conversão de custos_importacao.json para CSV

Este notebook lê `datasets/custos_importacao.json` e gera `datasets/custos_importacao_normalizado.csv` com o layout:

- `product_id` (integer)
- `product_name` (text)
- `category` (text)
- `start_date` (date)
- `usd_price` (float)

Objetivo: extrair o histórico aninhado `historic_data` em linhas normalizadas.

```python
import json
import csv
from pathlib import Path

json_path = Path('../datasets/custos_importacao.json')
csv_path = Path('../datasets/custos_importacao_normalizado.csv')
```

## Carregar JSON e extrair histórico

Ler o arquivo JSON e gerar uma lista de linhas normalizadas.

```python
with json_path.open('r', encoding='utf-8') as f:
    produtos = json.load(f)

linhas = []
for produto in produtos:
    product_id = produto.get('product_id')
    product_name = produto.get('product_name')
    category = produto.get('category')
    for entry in produto.get('historic_data', []):
        linhas.append({
            'product_id': product_id,
            'product_name': product_name,
            'category': category,
            'start_date': entry.get('start_date'),
            'usd_price': entry.get('usd_price')
        })

print(f'Total de linhas normalizadas: {len(linhas)}')
```

## Gravar CSV normalizado

Gerar o arquivo final em CSV com separador `,` e cabeçalho adequado.

```python
with csv_path.open('w', newline='', encoding='utf-8') as f:
    fieldnames = ['product_id', 'product_name', 'category', 'start_date', 'usd_price']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(linhas)

print(f'Arquivo salvo: {csv_path} (linhas: {len(linhas)})')
```

## Validação rápida

Ler de volta o CSV gerado e exibir as primeiras linhas para checagem.

```python
with csv_path.open('r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for i, row in enumerate(reader):
        if i >= 5:
            break
        print(row)
```

## Contagem total de entradas

Confirmando o número total de entradas no CSV normalizado.

```python
with csv_path.open('r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    total_entradas = sum(1 for row in reader)

print(f'Total de entradas no CSV normalizado: {total_entradas}')
```
