# Normalização de dados de vendas

Este notebook limpa e normaliza o arquivo `vendas_2023_2024.csv`, preparando-o para análises posteriores.

## Tarefas
- Carregar o CSV
- Normalizar datas (formato inconsistente)
- Remover linhas com datas inválidas
- Verificar e remover duplicatas
- Salvar CSV limpo

```python
import pandas as pd
```

## Carregar dados de vendas

```python
vendas = pd.read_csv('datasets/vendas_2023_2024.csv')
print(f"Linhas originais: {len(vendas)}")
print(vendas.head())
```

## Normalizar datas

```python
    # Função para analisar datas com múltiplos formatos
    def parse_date(date_str):
        # Tentar formato brasileiro (dd-mm-yyyy) primeiro
        try:
            return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce')
        except:
            pass
        # Tentar formato ISO (yyyy-mm-dd)
        try:
            return pd.to_datetime(date_str, format='%Y-%m-%d', errors='coerce')
        except:
            pass
        # Fallback: deixar pandas tentar inferir
        return pd.to_datetime(date_str, errors='coerce')
    
    vendas['sale_date'] = vendas['sale_date'].apply(parse_date)
    linhas_invalidas = vendas['sale_date'].isna().sum()
    print(f"Datas inválidas encontradas: {linhas_invalidas}")
    vendas = vendas.dropna(subset=['sale_date'])
    print(f"Linhas após remoção de datas inválidas: {len(vendas)}")
    print(vendas.head())
```

## Verificar e remover duplicatas

```python
duplicatas = vendas.duplicated().sum()
print(f"Duplicatas encontradas: {duplicatas}")
vendas = vendas.drop_duplicates()
print(f"Linhas após remoção de duplicatas: {len(vendas)}")
```

## Salvar CSV limpo

```python
vendas.to_csv('datasets/vendas_normalizado.csv', index=False)
print("Arquivo salvo: datasets/vendas_normalizado.csv")
```

- Datas inválidas removidas: 0
- Duplicatas removidas: 0
